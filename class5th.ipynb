```python
sns.distplot(a=data[data['Embarked']=="C"]["Survived"],bins=3,kde=False)
plt.xticks([0,1])
```

```python
sns.distplot(a=data[data['Embarked']=="Q"]['Survived'],kde=False,bins=3)
plt.xticks([0,1])
```

```python
figbi,axebi=plt.subplots(2,4,figsize=(18,10))
data.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axebi[0,0],xlim=[0,1])
data.groupby('SibSp')['Survived'].mean().plot(kind='barh',ax=axebi[0,1],xlim=[0,1])
data.groupby('Parch')['Survived'].mean().plot(kind='barh',ax=axebi[0,2],xlim=[0,1])
data.groupby('Sex')['Survived'].mean().plot(kind='barh',ax=axebi[1,0],xlim=[0,1])
data.groupby('Embarked')['Survived'].mean().plot(kind='barh',ax=axebi[1,1],xlim=[0,1])
sns.boxplot(x='Survived',y='Fare',data=data,ax=axebi[1,2])
plt.show()
```
