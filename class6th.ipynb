```python
beml=pd.read_csv('/content/BEML.csv',parse_dates=True)
glaxo=pd.read_csv('/content/GLAXO .csv',parse_dates=True)
beml.head()
```

```python
glaxo.head()
```

```python
glaxo.info()
```

```python
beml=beml[['Date','Close']]
glaxo=glaxo[['Date','Close']]
```

```python
beml=beml.set_index(pd.DatetimeIndex(beml['Date']))
glaxo=glaxo.set_index(pd.DatetimeIndex(glaxo['Date']))
```

```python
beml.info()
```

```python
beml.head()
```

```python
glaxo.head()
```

```python
%matplotlib inline
plt.plot(glaxo.Close)
```

```python
plt.plot(beml.Close)
```

```python
glaxo.gain=glaxo.Close.pct_change(periods=1)
```

```python
beml.gain=beml.Close.pct_change(periods=1)
```

```python
glaxo.dropna(inplace=True)
```

```python
plt.figure(figsize=(15,6))
plt.plot(beml.index,beml.gain)
```

```python
sns.distplot(beml.gain)
```

```python
sns.distplot(glaxo.gain,label="glaxo")
```

```python
plt.plot(glaxo.Close)
```
